```python
path = r'/kaggle/input/covid19/dataset.xlsx'
data = pd.read_excel(path)
#print(data.head())
```

```python
# Deleting this since it's all NaN
data = data.dropna(subset=['Patient ID', 'Patient age quantile', 'SARS-Cov-2 exam result', 'Inf A H1N1 2009', 'Respiratory Syncytial Virus', 'Sodium', 'Potassium', 'Mean corpuscular volume (MCV)', 'Proteina C reativa mg/dL'])
```

```python
#colunas
list(data.columns.values)
```

```python
covid_19 = pd.DataFrame(data, columns=['Patient ID', 'Patient age quantile', 'SARS-Cov-2 exam result', 'Inf A H1N1 2009', 'Respiratory Syncytial Virus', 'Sodium', 'Potassium', 'Mean corpuscular volume (MCV)', 'Proteina C reativa mg/dL']) 
covid_19
```

```python
from sklearn.preprocessing import LabelEncoder
lb_covid_19 = LabelEncoder()
covid_19["SARS-Cov-2 exam result_code"] = lb_covid_19.fit_transform(covid_19["SARS-Cov-2 exam result"])
#covid_19.sample(frac=.1).plot.scatter(x='Patient age quantile', y='SARS-Cov-2 exam result')
```
